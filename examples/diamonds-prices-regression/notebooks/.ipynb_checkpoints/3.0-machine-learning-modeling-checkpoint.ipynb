# Machine Learning Modeling Phase
In this first notebook I'm running the machine learning modeling phase for the Kaggle's <a href="https://www.kaggle.com/datasets/nancyalaswad90/diamonds-prices" target="_blank">diamonds price dataset</a>. This is a regression problem.

**Author**: Arthur G.

## Loading Dependencies
Here I'm loading and setting up the dependencies for this notebook.

```python
# libs
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pycaret.regression import *

# settings
seed = np.random.seed(42)
plt.style.use('ggplot')
```

## Helper Functions
A set of helper functions to automate data processing.

## Loading Dataset
Here I'm loading the dataset for initial processing.

```python
data = pd.read_csv(os.path.join('..', 'data', 'processed', 'processed_data.csv'))
data = data[['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z', 'price']]
data.head()
```

## Train-Test Split
Before the modeling phase, I'll be splitting the dataset into train and holdout set.

```python
# getting holdout data
holdout_set = data.sample(frac=0.1, random_state=seed)

# resetting data index
data = data.drop(index=holdout_set.index)
data.reset_index(drop=True)

data.head()
```

## Machine Learning Modeling

### Regression Experimento Setup
In this section I'm setting up the machine learning regression experiment with PyCaret.

```python
reg_exp = setup(
    data=data,
    target="price",
    train_size=0.8,
    ordinal_features={"cut" : ["Fair", "Good", "Very Good", "Ideal", "Premium"]},
    numeric_features=["carat", "depth", "table", "x", "y", "z"],
    categorical_features=["color", "clarity"],
    polynomial_features=True,
    normalize=True,
    remove_multicollinearity=True,
    transform_target=True,
    transform_target_method="box-cox",
    use_gpu=False
)
```

### Comparing Regression Models
In this next step I'm comparing different models' performances to pick the best one as our base model for further improvement..

```python
base_model = compare_models(sort="RMSE")
```

LightGBM was the model that outperformed all the other ones with a very decent RMSE as well as a high R2 score and an excelent Mean Absolute Percentage Error rate. 

Even though we have a pretty decent base model, in the next step I'll try to improve its performance with a model tuning algorithm. However, based on my experience I can say that we'll not achieve a great improvement as our base model already has pretty good performance indicators.

### Model Tuning
In this step I'll not make use of more advanced hyperparameter tuning algorithms such as Bayesian Optimization because it would consume too much time and computational power for no reason, as we have a not so big dataset as well as a really good base model.

```python
tuned_model = tune_model(base_model, optimize="RMSE")
```

As I anticipated above, we didn't achieve an improvement, in fact we even downgraded the performance of our base model. Our base model already had figured out the most optimized set of hyperparameters in the search space. By ignoring them and trying to find another set by trying to optimize our algorithm ever further ended up with a not so optimal set.

We can safely assume that trying to optimize our base model with different optimization strategies will keep this pattern of downgrading the models' performance.

### Model Plots
Now I'll be plotting a set of model's performance and healthy indicators collected during the training and validation time.

#### Model Plots: Residuals
A plot for the residuals in the training and testing sets.

```python
_ = plot_model(base_model, plot="residuals")
```

We can see that our set of optimized parameters did a really good job making the model to generalize well our data in both training and testing sets.

#### Model Plots: Error
A plot for the error between the **real y** and the  **predicted y**.

```python
_ = plot_model(base_model, plot="error")
```

Such a good performance!

#### Model Plots: Features
A plot for the features used for training the model.

```python
_ = plot_model(base_model, plot="feature_all")
```

The variables **carat**, **depth** and **cut** are the most important ones.

### Model Interpretation
It's time for us to get a sense of "how does the model works?" using a Shap-based interpretation plot.

```python
_ = interpret_model(base_model)
```

The most clear interpratation we can extract from this plot is "the higher the carat, the higher the diamond price".

### Model Validation
It's time to use both test and hould-out set's for the final validation of the model.

Test set validation.

```python
test_preds = predict_model(base_model)
```

Hold-out set validation.

```python
holdout_preds = predict_model(base_model, data=holdout_set)
```

The model's on both test and holdout sets is excellent, which shows us that the model is generalizing well.

### Model Finalization
Now we can finalize the model by training it with all the data available to make it ready to production.

```python
final_regressor = finalize_model(base_model)
```

### Model Serialization
Here I'm serializing the finalized model to a ".pkl" file.

```python
save_model(final_regressor, os.path.join("..", "results", "models", "lightgbm_price_regressor"), verbose=False)
```
